```python
#Importar
from bs4 import BeautifulSoup
import requests

import pandas as pd
import re
```

```python
def datos_TopArtista(url):

    TopArtista_bolsa = requests.get(url)
    print(f"La respuesta a la peticion es: {TopArtista_bolsa.status_code}")

    sopa_TopArtista = BeautifulSoup(TopArtista_bolsa.content, "html.parser")
    tablas = sopa_TopArtista.find_all("table")
    nuestra_tabla = tablas[0]

    encabezados_TopArtista = ["Posición", "P+", "Artista y canción", "Dias", "Posición más alta (xdias)", "Streams", "Streams+", "7Day", "7Day+", "Total"]
    filas_TopArtista = nuestra_tabla.find_all("tr")

    resultados_TopArtista = []
    for fila in filas_TopArtista[1:]:
        resultados_TopArtista.append(fila.text.split("\n")[0:10])

    for indice_lista, lista in enumerate(resultados_TopArtista):
        for indice_elemento, elemento in enumerate(lista):
            try:
                resultados_TopArtista[indice_lista][indice_elemento] = elemento.replace('\xa0', '').strip()
            except:
                resultados_TopArtista[indice_lista][indice_elemento] = elemento

    # Crear el DataFrame con los resultados
    df_TopArtista = pd.DataFrame(resultados_TopArtista, columns=encabezados_TopArtista)

    return df_TopArtista
```

```python

# Código para ALBUM y obtener su DATAFRAME

artistas_list = ['Ke Personajes', 'Ariana Grande', 'Linkin Park', 'Amelie lens', 'Maneskin']

album_data = []

for artista_nombre in tqdm.tqdm(artistas_list, desc='Obteniendo álbumes de artistas:'):
    resultados = sp.search(q=f'artist:{artista_nombre}', type='album')

    for album in resultados['albums']['items']:
        nombre_album = album['name']
        fecha_lanzamiento = album['release_date']

        artistas = [artist['name'] for artist in album['artists']]
        artista_principal = artistas[0] if artistas else "Artista no disponible"

        genero = None
        artista_id = None  # Inicializa artista_id fuera del bloque if
        
        if album['artists']:
            artista_id = album['artists'][0]['id']  # Ahora artista_id tiene el ID del artista
            artista_info = sp.artist(artista_id)
            generos_artista = artista_info.get('genres', [])
            if generos_artista:
                genero = generos_artista[0]

        año_lanzamiento = fecha_lanzamiento[:4]  # Obtiene solo el año

        # Verifica si el año está en el rango deseado
        if 2019 <= int(año_lanzamiento) <= 2024:
            album_info = {
                "ID": artista_id,  # Usa artista_id en lugar de album['id']
                "Artista": artista_principal,
                "Género": genero,
                "Tipo": "Álbum",
                "Nombre_album": nombre_album,
                "Año_lanzamiento": año_lanzamiento
            }

            album_data.append(album_info)

df_albumes = pd.DataFrame(album_data)
df_albumes #Esta data es un documento virtual que se introduce al SQL
```
